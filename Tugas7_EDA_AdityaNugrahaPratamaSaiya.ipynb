**2. Analisis Distribusi**

> Plot distribusi TotalCases, TotalDeaths, dan TotalRecovered

> Buat histogram log-skala untuk kolom TotalCases

> Bandingkan populasi terhadap jumlah kasus (scatter plot)

```python
# Plot Distribusi TotalCases, TotalDeaths, TotalRecovered
# Menentukan kolom yang ingin dianalisis distribusinya
columns = ['TotalCases', 'TotalDeaths', 'TotalRecovered']
labels  = ['Total Cases', 'Total Deaths', 'Total Recovered']
colors  = ['steelblue', 'tomato', 'seagreen']

# Membuat grid 3 baris x 3 kolom untuk 9 plot sekaligus
fig, axes = plt.subplots(3, 3, figsize=(18, 13))
fig.suptitle('Distribusi TotalCases, TotalDeaths & TotalRecovered', fontsize=15, fontweight='bold', y=1.01)

for i, (col, label, color) in enumerate(zip(columns, labels, colors)):

    # Histogram + KDE (kurva distribusi)
    ax1 = axes[i][0]
    sns.histplot(df[col].dropna(), kde=True, color=color, ax=ax1, bins=30)
    # Garis merah putus-putus = nilai rata-rata
    ax1.axvline(df[col].mean(),   color='red',    linestyle='--', linewidth=1.5,
                label=f'Mean: {df[col].mean():,.0f}')
    # Garis oranye titik-titik = nilai tengah (median)
    ax1.axvline(df[col].median(), color='orange', linestyle=':',  linewidth=1.5,
                label=f'Median: {df[col].median():,.0f}')
    ax1.set_title(f'{label} - Histogram + KDE')
    ax1.set_xlabel(label)
    ax1.set_ylabel('Frekuensi')
    ax1.legend(fontsize=8)

    # Box Plot (melihat outlier dan sebaran data)
    ax2 = axes[i][1]
    sns.boxplot(y=df[col].dropna(), color=color, ax=ax2)
    ax2.set_title(f'{label} - Box Plot')
    ax2.set_ylabel(label)

    # Violin Plot (gabungan KDE dan Box Plot)
    ax3 = axes[i][2]
    sns.violinplot(y=df[col].dropna(), color=color, ax=ax3, inner='quartile')
    ax3.set_title(f'{label} - Violin Plot')
    ax3.set_ylabel(label)

plt.tight_layout()
plt.show()
```

```python
# Histogram Log-Skala TotalCases

# Perbandingan skala normal vs log
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('TotalCases: Skala Normal vs Log', fontsize=13, fontweight='bold')

# Skala normal: terlihat sangat tidak merata (skewed ke kanan)
axes[0].hist(df['TotalCases'].dropna(), bins=40, color='steelblue', edgecolor='white')
axes[0].set_title('Skala Normal')
axes[0].set_xlabel('Total Cases')
axes[0].set_ylabel('Frekuensi')

# Skala log pada sumbu Y: distribusi lebih mudah dibaca
axes[1].hist(df['TotalCases'].dropna(), bins=40, color='steelblue', edgecolor='white', log=True)
axes[1].set_title('Skala Log (sumbu Y)')
axes[1].set_xlabel('Total Cases')
axes[1].set_ylabel('Frekuensi (log)')

plt.tight_layout()
plt.show()

# Distribusi log10(TotalCases): transformasi nilai ke skala logaritma
fig2, ax = plt.subplots(figsize=(8, 5))

# np.log10 mengubah angka besar menjadi skala lebih kecil agar mudah dibaca
log_cases = np.log10(df['TotalCases'].replace(0, np.nan).dropna())

sns.histplot(log_cases, kde=True, color='steelblue', ax=ax, bins=25)
ax.axvline(log_cases.mean(),   color='red',    linestyle='--', linewidth=1.5,
           label=f'Mean log10: {log_cases.mean():.2f}')
ax.axvline(log_cases.median(), color='orange', linestyle=':',  linewidth=1.5,
           label=f'Median log10: {log_cases.median():.2f}')
ax.set_title('Distribusi log10(TotalCases)')
ax.set_xlabel('log10(TotalCases)')
ax.set_ylabel('Frekuensi')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Scatter Plot: Populasi vs TotalCases

# Hapus baris yang memiliki nilai kosong pada kolom yang dibutuhkan
df_scatter = df[['Country/Region', 'Population', 'TotalCases']].dropna()

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Populasi vs Jumlah Kasus COVID-19', fontsize=13, fontweight='bold')

# Skala Normal
ax1 = axes[0]
ax1.scatter(df_scatter['Population'], df_scatter['TotalCases'],
            alpha=0.6, color='steelblue', edgecolors='white', s=60)

# Beri label nama untuk 5 negara dengan kasus terbanyak
top5 = df_scatter.nlargest(5, 'TotalCases')
for _, row in top5.iterrows():
    ax1.annotate(row['Country/Region'],
                 xy=(row['Population'], row['TotalCases']),
                 xytext=(5, 5), textcoords='offset points',
                 fontsize=8, color='darkred')
ax1.set_title('Skala Normal')
ax1.set_xlabel('Populasi')
ax1.set_ylabel('TotalCases')

# Skala Log-Log (lebih jelas melihat pola)
ax2 = axes[1]
log_pop   = np.log10(df_scatter['Population'].replace(0, np.nan))
log_cases = np.log10(df_scatter['TotalCases'].replace(0, np.nan))

ax2.scatter(log_pop, log_cases, alpha=0.6, color='tomato',
            edgecolors='white', s=60)

# Tambahkan garis tren (regresi linear)
mask = log_pop.notna() & log_cases.notna()
z = np.polyfit(log_pop[mask], log_cases[mask], 1)
p = np.poly1d(z)
x_line = np.linspace(log_pop[mask].min(), log_pop[mask].max(), 100)
ax2.plot(x_line, p(x_line), color='black', linestyle='--', linewidth=1.5,
         label=f'Tren (slope={z[0]:.2f})')

# Label top 5 negara
for _, row in top5.iterrows():
    ax2.annotate(row['Country/Region'],
                 xy=(np.log10(row['Population']), np.log10(row['TotalCases'])),
                 xytext=(5, 5), textcoords='offset points',
                 fontsize=8, color='darkred')
ax2.set_title('Skala Log-Log')
ax2.set_xlabel('log10(Populasi)')
ax2.set_ylabel('log10(TotalCases)')
ax2.legend()

plt.tight_layout()
plt.show()

# Tampilkan nilai korelasi antara populasi dan jumlah kasus
corr = df_scatter['Population'].corr(df_scatter['TotalCases'])
print(f'Korelasi Populasi vs TotalCases: {corr:.4f}')
```

**3. Analisis Per Benua (Continent)**

> Hitung total kasus, kematian, dan kesembuhan per benua

> Visualisasikan perbandingan kasus dan kematian antar benua

```python
# Mengelompokkan data berdasarkan Continent dan menjumlahkan tiap kolom
continent_df = df.groupby('Continent')[['TotalCases', 'TotalDeaths', 'TotalRecovered', 'ActiveCases']].sum()
continent_df = continent_df.sort_values('TotalCases', ascending=False).reset_index()

print('Total Kasus, Kematian, dan Kesembuhan per Benua:')
print(continent_df.to_string(index=False))

# Hitung Death Rate dan Recovery Rate per benua (dalam persen)
continent_df['Death Rate (%)']    = (continent_df['TotalDeaths']    / continent_df['TotalCases'] * 100).round(2)
continent_df['Recovery Rate (%)'] = (continent_df['TotalRecovered'] / continent_df['TotalCases'] * 100).round(2)

print()
print('Death Rate & Recovery Rate per Benua:')
print(continent_df[['Continent', 'Death Rate (%)', 'Recovery Rate (%)']].to_string(index=False))
```

```python
# Visualisasi perbandingan kasus, kematian, dan kesembuhan per benua
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Analisis COVID-19 per Benua', fontsize=15, fontweight='bold')

colors = sns.color_palette('Set2', len(continent_df))

# Total Cases per Benua
ax1 = axes[0][0]
bars1 = ax1.barh(continent_df['Continent'], continent_df['TotalCases'], color=colors)
ax1.set_title('Total Cases per Benua', fontweight='bold')
ax1.set_xlabel('Total Cases')
ax1.bar_label(bars1, labels=[f'{v:,.0f}' for v in bars1.datavalues], padding=3, fontsize=8)
ax1.invert_yaxis()

# Total Deaths per Benua
ax2 = axes[0][1]
bars2 = ax2.barh(continent_df['Continent'], continent_df['TotalDeaths'], color='tomato')
ax2.set_title('Total Deaths per Benua', fontweight='bold')
ax2.set_xlabel('Total Deaths')
ax2.bar_label(bars2, labels=[f'{v:,.0f}' for v in bars2.datavalues], padding=3, fontsize=8)
ax2.invert_yaxis()

# Grouped Bar: Cases, Deaths, Recovered sekaligus
ax3 = axes[1][0]
x     = range(len(continent_df))
width = 0.25
ax3.bar([i - width for i in x], continent_df['TotalCases'],     width=width, label='Cases',     color='steelblue')
ax3.bar([i         for i in x], continent_df['TotalDeaths'],    width=width, label='Deaths',    color='tomato')
ax3.bar([i + width for i in x], continent_df['TotalRecovered'], width=width, label='Recovered', color='seagreen')
ax3.set_title('Perbandingan Cases, Deaths & Recovered', fontweight='bold')
ax3.set_xticks(list(x))
ax3.set_xticklabels(continent_df['Continent'], rotation=20, ha='right', fontsize=9)
ax3.set_ylabel('Jumlah')
ax3.legend()

# Death Rate vs Recovery Rate per Benua
ax4 = axes[1][1]
ax4.bar([i - 0.2 for i in x], continent_df['Death Rate (%)'],    width=0.35, label='Death Rate (%)',    color='tomato')
ax4.bar([i + 0.2 for i in x], continent_df['Recovery Rate (%)'], width=0.35, label='Recovery Rate (%)', color='seagreen')
ax4.set_title('Death Rate & Recovery Rate per Benua', fontweight='bold')
ax4.set_xticks(list(x))
ax4.set_xticklabels(continent_df['Continent'], rotation=20, ha='right', fontsize=9)
ax4.set_ylabel('Rate (%)')
ax4.legend()

plt.tight_layout()
plt.show()
```

**4. Analisis Tes dan Positivity Rate**
> Buat kolom baru "Positivity Rate" = TotalCases / TotalTests

> Analisis negara-negara dengan positivity rate tertinggi

```python
# Salin dataframe agar data asli tidak berubah
df_test = df[['Country/Region', 'Continent', 'TotalCases', 'TotalTests', 'Tests/1M pop']].dropna().copy()

# Hitung Positivity Rate: berapa persen tes yang hasilnya positif
df_test['Positivity Rate (%)'] = (df_test['TotalCases'] / df_test['TotalTests'] * 100).round(2)

# Urutkan dari positivity rate tertinggi
df_test = df_test.sort_values('Positivity Rate (%)', ascending=False).reset_index(drop=True)

print(f'Jumlah negara dengan data tes lengkap: {len(df_test)}')
print()
print('Statistik Positivity Rate global:')
print(df_test['Positivity Rate (%)'].describe().round(2))
print()

# Tampilkan 10 negara dengan positivity rate tertinggi
print('=== Top 10 Negara - Positivity Rate Tertinggi ===')
print(df_test[['Country/Region', 'TotalCases', 'TotalTests', 'Positivity Rate (%)']].head(10).to_string(index=False))
```

```python
# Visualisasi Positivity Rate COVID-19
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Analisis Positivity Rate COVID-19', fontsize=14, fontweight='bold')

# Top 15 Negara Positivity Rate Tertinggi
ax1 = axes[0]
top15 = df_test.nlargest(15, 'Positivity Rate (%)')
bars = ax1.barh(top15['Country/Region'], top15['Positivity Rate (%)'], color='coral')
ax1.set_title('Top 15 Negara - Positivity Rate Tertinggi', fontweight='bold')
ax1.set_xlabel('Positivity Rate (%)')
ax1.bar_label(bars, fmt='%.1f%%', padding=3, fontsize=8)
ax1.invert_yaxis()

# Scatter: TotalTests vs TotalCases dengan warna berdasarkan Positivity Rate
ax2 = axes[1]
scatter = ax2.scatter(df_test['TotalTests'], df_test['TotalCases'],
                      c=df_test['Positivity Rate (%)'],
                      cmap='YlOrRd', alpha=0.7, s=80, edgecolors='white')

# Colorbar sebagai legenda warna
plt.colorbar(scatter, ax=ax2, label='Positivity Rate (%)')

# Label top 5 negara
top5 = df_test.nlargest(5, 'TotalCases')
for _, row in top5.iterrows():
    ax2.annotate(row['Country/Region'],
                 xy=(row['TotalTests'], row['TotalCases']),
                 xytext=(5, 5), textcoords='offset points',
                 fontsize=8, color='darkred')
ax2.set_title('TotalTests vs TotalCases\n(warna = Positivity Rate)', fontweight='bold')
ax2.set_xlabel('Total Tests')
ax2.set_ylabel('Total Cases')

plt.tight_layout()
plt.show()
```

**5. Death Rate & Recovery Rate**
> Buat kolom baru "Death Rate" = TotalDeaths / TotalCases

> Buat kolom "Recovery Rate" = TotalRecovered / TotalCases

> Bandingkan death rate dan recovery rate secara global

```python
# Salin kolom yang diperlukan dan hapus baris kosong
df_rate = df[['Country/Region', 'Continent', 'TotalCases', 'TotalDeaths', 'TotalRecovered']].dropna().copy()

# Hitung Death Rate: persentase kematian dari total kasus
df_rate['Death Rate (%)']    = (df_rate['TotalDeaths']    / df_rate['TotalCases'] * 100).round(2)

# Hitung Recovery Rate: persentase kesembuhan dari total kasus
df_rate['Recovery Rate (%)'] = (df_rate['TotalRecovered'] / df_rate['TotalCases'] * 100).round(2)

print('Statistik Death Rate global:')
print(df_rate['Death Rate (%)'].describe().round(2))
print()
print('Statistik Recovery Rate global:')
print(df_rate['Recovery Rate (%)'].describe().round(2))
```

```python
# Top 10 negara dengan Death Rate tertinggi
top10_death = df_rate.nlargest(10, 'Death Rate (%)')[['Country/Region', 'TotalCases', 'TotalDeaths', 'Death Rate (%)']]
print('=== Top 10 Negara - Death Rate Tertinggi ===')
print(top10_death.to_string(index=False))
print()

# Top 10 negara dengan Recovery Rate tertinggi
top10_recovery = df_rate.nlargest(10, 'Recovery Rate (%)')[['Country/Region', 'TotalCases', 'TotalRecovered', 'Recovery Rate (%)']]
print('=== Top 10 Negara - Recovery Rate Tertinggi ===')
print(top10_recovery.to_string(index=False))
```

```python
# Visualisasi perbandingan Death Rate dan Recovery Rate secara global
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Perbandingan Death Rate & Recovery Rate Global', fontsize=14, fontweight='bold')

# Top 10 negara Death Rate tertinggi
ax1 = axes[0]
top10_d = df_rate.nlargest(10, 'Death Rate (%)')
bars1 = ax1.barh(top10_d['Country/Region'], top10_d['Death Rate (%)'], color='tomato')
ax1.set_title('Top 10 Negara - Death Rate Tertinggi', fontweight='bold')
ax1.set_xlabel('Death Rate (%)')
ax1.bar_label(bars1, fmt='%.2f%%', padding=3, fontsize=8)
ax1.invert_yaxis()

# Top 10 negara Recovery Rate tertinggi
ax2 = axes[1]
top10_r = df_rate.nlargest(10, 'Recovery Rate (%)')
bars2 = ax2.barh(top10_r['Country/Region'], top10_r['Recovery Rate (%)'], color='seagreen')
ax2.set_title('Top 10 Negara - Recovery Rate Tertinggi', fontweight='bold')
ax2.set_xlabel('Recovery Rate (%)')
ax2.bar_label(bars2, fmt='%.1f%%', padding=3, fontsize=8)
ax2.invert_yaxis()

# Scatter: Death Rate vs Recovery Rate per negara (warna per benua)
ax3 = axes[2]
continents = df_rate['Continent'].dropna().unique()
palette = sns.color_palette('Set2', len(continents))
for continent, color in zip(continents, palette):
    subset = df_rate[df_rate['Continent'] == continent]
    ax3.scatter(subset['Death Rate (%)'], subset['Recovery Rate (%)'],
                label=continent, color=color, alpha=0.7, s=60, edgecolors='white')
ax3.set_title('Death Rate vs Recovery Rate per Negara', fontweight='bold')
ax3.set_xlabel('Death Rate (%)')
ax3.set_ylabel('Recovery Rate (%)')
ax3.legend(fontsize=7)

plt.tight_layout()
plt.show()
```

- Rata-rata Recovery Rate global sebesar **68.82%**

#### Top 10 Negara Recovery Rate Tertinggi

**6. Serious/Critical Case Analysis**

> Visualisasi negara dengan jumlah kasus kritis tertinggi

> Bandingkan dengan jumlah kasus aktif

```python
# Kolom 'Serious,Critical' berisi jumlah pasien dalam kondisi kritis
df_critical = df[['Country/Region', 'Continent', 'TotalCases', 'ActiveCases', 'Serious,Critical']].dropna().copy()

# Hitung persentase kasus kritis dari kasus aktif
df_critical['Critical Rate (%)'] = (df_critical['Serious,Critical'] / df_critical['ActiveCases'] * 100).round(2)

print('Statistik kasus Serious/Critical global:')
print(df_critical['Serious,Critical'].describe().round(0))
print()
print('=== Top 10 Negara - Jumlah Kasus Kritis Terbanyak ===')
top10_crit = df_critical.nlargest(10, 'Serious,Critical')[['Country/Region', 'ActiveCases', 'Serious,Critical', 'Critical Rate (%)']]
print(top10_crit.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Analisis Kasus Serious/Critical COVID-19', fontsize=14, fontweight='bold')

# Top 15 negara dengan kasus kritis terbanyak
ax1 = axes[0]
top15_c = df_critical.nlargest(15, 'Serious,Critical')
bars = ax1.barh(top15_c['Country/Region'], top15_c['Serious,Critical'], color='darkorange')
ax1.set_title('Top 15 Negara - Kasus Serious/Critical', fontweight='bold')
ax1.set_xlabel('Jumlah Kasus Kritis')
ax1.bar_label(bars, labels=[f'{v:,.0f}' for v in bars.datavalues], padding=3, fontsize=8)
ax1.invert_yaxis()

# Scatter: ActiveCases vs Serious/Critical
ax2 = axes[1]
ax2.scatter(df_critical['ActiveCases'], df_critical['Serious,Critical'],
            alpha=0.6, color='darkorange', edgecolors='white', s=60)

# Label top 5 negara
top5_c = df_critical.nlargest(5, 'Serious,Critical')
for _, row in top5_c.iterrows():
    ax2.annotate(row['Country/Region'],
                 xy=(row['ActiveCases'], row['Serious,Critical']),
                 xytext=(5, 5), textcoords='offset points',
                 fontsize=8, color='darkred')

# Garis tren
z = np.polyfit(df_critical['ActiveCases'], df_critical['Serious,Critical'], 1)
x_line = np.linspace(df_critical['ActiveCases'].min(), df_critical['ActiveCases'].max(), 100)
ax2.plot(x_line, np.poly1d(z)(x_line), 'k--', linewidth=1.5, label='Tren')
ax2.set_title('Active Cases vs Serious/Critical', fontweight='bold')
ax2.set_xlabel('Active Cases')
ax2.set_ylabel('Serious/Critical')
ax2.legend()

plt.tight_layout()
plt.show()
```

**8. Correlation Matrix**

> Buat heatmap korelasi untuk kolom numerik

> Apa yang menarik dari korelasi antara populasi, test, dan kasus?

```python
# Pilih hanya kolom numerik yang relevan untuk analisis korelasi
kolom_numerik = ['TotalCases', 'TotalDeaths', 'TotalRecovered',
                 'ActiveCases', 'TotalTests', 'Population',
                 'Serious,Critical', 'Tests/1M pop']

# Hitung matriks korelasi (nilai -1 s/d 1)
# Nilai mendekati 1  = korelasi positif kuat
# Nilai mendekati -1 = korelasi negatif kuat
# Nilai mendekati 0  = tidak ada korelasi
corr_matrix = df[kolom_numerik].corr().round(2)

print('Matriks Korelasi:')
print(corr_matrix)
```

```python
# Heatmap: warna merah = korelasi positif, biru = korelasi negatif
# annot=True: tampilkan angka di tiap kotak
# fmt='.2f': format 2 desimal
# cmap='coolwarm': warna merah-putih-biru
# square=True: kotak berbentuk persegi
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='coolwarm',
            square=True,
            linewidths=0.5,
            ax=ax)

ax.set_title('Heatmap Korelasi Antar Variabel COVID-19', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Insight - Correlation Matrix

#### Korelasi Sangat Kuat (> 0.90)
- **TotalCases dan TotalRecovered** memiliki korelasi tertinggi (**0.99**),
  artinya semakin banyak kasus maka semakin banyak yang sembuh
- **TotalCases dan Serious,Critical** berkorelasi **0.97**,
  menunjukkan kasus kritis sangat dipengaruhi oleh jumlah total kasus
- **TotalCases dan ActiveCases** berkorelasi **0.97**,
  wajar karena kasus aktif merupakan bagian dari total kasus

#### Korelasi Sedang (0.40 - 0.90)
- **Population dan TotalCases** hanya berkorelasi **0.55**,
  artinya populasi besar tidak selalu berarti kasus banyak
- **TotalTests dan TotalCases** berkorelasi **0.89**,
  negara dengan lebih banyak tes cenderung menemukan lebih banyak kasus

#### Korelasi Lemah / Tidak Ada (< 0.15)
- **Tests/1M pop** memiliki korelasi sangat rendah terhadap semua variabel,
  bahkan **negatif terhadap Population (-0.08)**,
  artinya negara berpopulasi besar justru cenderung memiliki rasio tes per kapita yang lebih rendah
- **Tests/1M pop dan TotalCases** hanya **0.03**,
  menunjukkan bahwa banyaknya tes per kapita tidak berkaitan langsung dengan jumlah kasus absolut
